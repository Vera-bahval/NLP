Данные разделены на три части: обучающую, валидационную и тестовую. Запишем их в соответствующие переменные.

Укажем директорию, куда сохранились признаки. Будем использовать русскоязычный BERT, что указано в переменной model_dir. Признаки находятся в папке features.

```python
model_dir = "/content/tda-bert-ru-cola"
heads = 16 if "roberta" in model_dir.lower() else 12
layers = 24 if "roberta" in model_dir.lower() else 12
features_dir = model_dir + "/features/"
```

Для загрузки признаков используется функция load_features, которая была определена при скачивании файла read_features.py из гитхаб-репозитория.

```python
kwargs ={"features_dir": features_dir,
         "model_dir": model_dir,
         "heads": heads,
         "layers": layers}
X_train, X_valid, X_test = list(map(lambda x: load_features(x, **kwargs), [x for x in (train_set_name, valid_set_name, test_set_name)]))
```

**2. Обучение модели логистической регрессии, подбор параметров с помощью GridSearch, оценка качества классификации.**

**Классификация включает в себя три этапа:**

*   стандартизация данных (метод StandardScaler)
*   уменьшение размерности признаковых векторов (метод главных компонент PCA)
*  обучение модели для классификации (логистическая регрессия)

Все этапы обработки объединяются в **Pipeline**.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

model = Pipeline([
    ('scaler', StandardScaler()),
    ('reduce_dim', PCA(random_state=42,
                       n_components=250)),
    ('clf', LogisticRegression(random_state=42,
                               solver='liblinear',
                               penalty='l1',
                               C=0.1))])
```

Собственно обучение модели

```python
model.fit(X_train, y_train)
```

Функция для вывода метрик

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import matthews_corrcoef

def print_scores(y, y_pred):
    acc, mcc = accuracy_score(y, y_pred), matthews_corrcoef(y, y_pred)
    print('Accuracy =', round(acc, 2))
    print('MCC =', round(mcc, 2))
```

Определим качество классификации для валидационной и тестовой выборки.

```python
print_scores(model.predict(X_valid), y_valid)
```

```python
print_scores(model.predict(X_test), y_test)
```

Можно заметить, что для валидационной выборки качество выше, чем для тестовой, а также, что значение коэффициента MMC почти одинаковое, однако гораздо ниже, чем значение Accuracy в обоих случаях.

Теперь воспользуемся GridSearch'ем для подбора лучших параметров обучения.

```python
from sklearn.model_selection import GridSearchCV
```

Для начала посмотрим, какие параметры есть у нашей модели.

```python
model.get_params()
```

Будем изменять значения параметров 'reduce_dim__n_components'и 'clf__C'. param_grid - словарь с параметрами и их значениями, которые хотим подбирать, cv - количество разрезов кросс-валидации, scoring - метрика, показатели которой хотим улучшить

```python
param_grid = {'reduce_dim__n_components': [50, 100, 150, 200, 250],  'clf__C': [0.01, 0.1, 1.0]  }
search = GridSearchCV(model, param_grid, cv=5, scoring="accuracy")
```

Проводим обучение

```python
search.fit(X_train, y_train)
```

Выводим лучшие параметры

```python
search.best_params_
```

Посмотрим, как изменилось качество классификации

```python
print_scores(search.predict(X_valid), y_valid)
```

```python
print_scores(search.predict(X_test), y_test)
```

С помощью GridSearch удалось установить, что наиболее оптимальное количество компонент в PCA = 100. Однако на качество это несильно повлияло. По факту, удалось немного улучшить показатель MCC для тестовой выборки.
